```python
import neuralnetwork as nn

import matplotlib.pyplot as plt
import numpy as np
  
x = np.linspace(-10, 10, 100)
z = 1/(1 + np.exp(-x))
y = nn.sigmoid(x)
g = nn.sigmoid_grad(x)
  
plt.plot(x, z, label="ext")
plt.plot(x,y, label="my")
plt.plot(x,g, label="grad")
plt.xlabel("x")
plt.ylabel("Sigmoid(X)")
plt.legend()
plt.grid(True)
plt.show()

y = nn.relu(x)
zero = np.zeros(len(x))
z = np.max([zero, x], axis=0)
g = nn.relu_grad(x)

plt.plot(x, z, label="ext")
plt.plot(x,y, label="my")
plt.plot(x,g, label="grad")
plt.xlabel("x")
plt.ylabel("Relu(X)")
plt.legend()
plt.grid(True)
plt.show()

y = nn.leaky_relu(x)
g = nn.leaky_relu_grad(x)

plt.plot(x,y, label="my")
plt.plot(x,g, label="grad")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()

y = nn.tanh(x)
g = nn.tanh_grad(x)

plt.plot(x,y, label="my")
plt.plot(x,g, label="grad")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()

y = nn.elu(x, alpha = 1)
g = nn.elu_grad(x, alpha = 1)

plt.plot(x,y, label="my")
plt.plot(x,g, label="grad")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```
